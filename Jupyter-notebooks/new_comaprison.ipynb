```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# Load all metric history files
# ============================================================

paths_fedavg = {
    "IID": "metrics_history_fedavg_iid_pretrained.npy",
    "Moderate": "metrics_history_fedavg_moderate_pretrained_alpha.npy",
    "Extreme": "metrics_history_fedavg_extreme_pretrained.npy"
}

paths_fedprox = {
    "IID": "metrics_history_fedprox_pretrained_iid.npy",
    "Moderate": "metrics_history_fedprox_moderate_pretrained_alpha.npy",
    "Extreme": "metrics_history_fedprox_extreme_pretrained_newmu1.npy"
}

def load_metrics(path):
    hist = np.load(path, allow_pickle=True).item()
    rounds = hist["round"]
    acc = [m["acc"] for m in hist["holdout"]]
    loss = [m["loss"] for m in hist["holdout"]]
    return rounds, acc, loss


# ============================================================
# Load FEDAVG metrics
# ============================================================

fedavg_data = {}
for scenario, path in paths_fedavg.items():
    rounds, acc, loss = load_metrics(path)
    fedavg_data[scenario] = {
        "rounds": rounds,
        "acc": acc,
        "loss": loss
    }


# ============================================================
# Load FEDPROX metrics
# ============================================================

fedprox_data = {}
for scenario, path in paths_fedprox.items():
    rounds, acc, loss = load_metrics(path)
    fedprox_data[scenario] = {
        "rounds": rounds,
        "acc": acc,
        "loss": loss
    }


# ============================================================
# PLOT 1 — FEDAVG Accuracy Across All Scenarios
# ============================================================

plt.figure(figsize=(10,5))

for scenario, info in fedavg_data.items():
    plt.plot(info["rounds"], info["acc"], marker='o', label=f"FedAvg {scenario}")

plt.title("FedAvg — Accuracy Across IID, Moderate, Extreme")
plt.xlabel("Rounds")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# ============================================================
# PLOT 2 — FEDPROX Accuracy Across All Scenarios
# ============================================================

plt.figure(figsize=(10,5))

for scenario, info in fedprox_data.items():
    plt.plot(info["rounds"], info["acc"], marker='o', label=f"FedProx {scenario}")

plt.title("FedProx — Accuracy Across IID, Moderate, Extreme")
plt.xlabel("Rounds")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print("✔ All plots generated successfully!")
```
